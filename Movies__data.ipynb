```python
# APP Building using .py file
import gradio as gr
import pandas as pd
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
import nltk

# Download required NLTK resources
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

# Load dataset (Ensure 'Train.csv' exists in the correct path)
df = pd.read_csv("/content/Train.csv", encoding='latin-1')

# Preprocessing function
def preprocess_text(text):
    stop_words = set(stopwords.words('english'))
    lemmatizer = WordNetLemmatizer()

    tokens = word_tokenize(str(text))  # Convert to string (for safety)
    cleaned = [word.lower() for word in tokens if word.isalpha() and word.lower() not in stop_words]
    lemmatized = [lemmatizer.lemmatize(word) for word in cleaned]
    return " ".join(lemmatized)

# Apply preprocessing
df["cleaned_text"] = df["ÿtext"].apply(preprocess_text)

# Initialize and fit TF-IDF Vectorizer
tf = TfidfVectorizer(lowercase=True, stop_words='english', max_df=0.95, max_features=1200)
X_train_tfidf = tf.fit_transform(df["cleaned_text"])  # **Fit on training data**
y_train = df["label"]

# Train the Naive Bayes model
mnb = MultinomialNB()
mnb.fit(X_train_tfidf, y_train)  # **Train model**

# Prediction function for Gradio
def classify_text(input_text):
    """Classify user input using the trained model."""
    processed_text = preprocess_text(input_text)
    transformed_text = tf.transform([processed_text])  # Now `tf` is fitted!
    prediction = mnb.predict(transformed_text.toarray())[0]
    return f"Predicted Label: {prediction}"

# Gradio UI
iface = gr.Interface(
    fn=classify_text,
    inputs=gr.Textbox(placeholder="Enter your text here..."),
    outputs=gr.Textbox(),
    title="Text Classification with Naive Bayes",
    description="Enter a sentence, and the model will predict its category."
)

# Launch app
iface.launch(share=True, debug=True)
```
