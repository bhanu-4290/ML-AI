```python
import gradio as gr
import pandas as pd
from nltk.tokenize import word_tokenize
from string import punctuation
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
import nltk
import warnings
warnings.simplefilter('ignore')

# Download required NLTK resources
nltk.download('punkt')

# Load dataset
data = pd.read_csv("Language_Detection_Dataset.csv")

# Splitting dataset into training and testing
x_train, x_test, y_train, y_test = train_test_split(data.Text, data.language, test_size=0.2, random_state=42)

# Encode language labels into numerical values
label_encoder = LabelEncoder()
y_train_encoded = label_encoder.fit_transform(y_train)
y_test_encoded = label_encoder.transform(y_test)

# Text preprocessing function
def preprocess(text):
    tokens = word_tokenize(text)
    tokens = [word for word in tokens if word not in punctuation]
    return " ".join(tokens)

# Apply preprocessing to training and testing text data
x_train_processed = x_train.apply(preprocess)
x_test_processed = x_test.apply(preprocess)

# TF-IDF Vectorization
vectorizer = TfidfVectorizer(lowercase=True, stop_words='english', max_df=0.95, max_features=1000)
x_train_tfidf = vectorizer.fit_transform(x_train_processed).toarray()
x_test_tfidf = vectorizer.transform(x_test_processed).toarray()

# Train Naive Bayes Model
model = MultinomialNB()
model.fit(x_train_tfidf, y_train_encoded)

# Prediction function for Gradio
def predict_language(text):
    processed_text = preprocess(text)  # Preprocess input text
    transformed_text = vectorizer.transform([processed_text]).toarray()  # Convert to TF-IDF
    prediction = model.predict(transformed_text)  # Predict language
    language = label_encoder.inverse_transform(prediction)[0]  # Convert label to language name
    return f"Predicted Language: {language}"  # Return prediction

# Gradio Interface
iface = gr.Interface(
    fn=predict_language,
    inputs=gr.Textbox(placeholder="Enter a sentence..."),
    outputs=gr.Textbox(),
    title="Language Identifier",
    description="Enter a sentence, and the model will predict its language."
)

# Launch Gradio app
iface.launch(share=True)
```
